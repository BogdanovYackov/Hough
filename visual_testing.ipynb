```python
filename='test_images/1.png'
```

```python
from PIL import Image
import numpy as np
from hough import hough, create_template, show

img = Image.open(filename)
arr = 1 - np.asarray(img)[..., 0] / 255
show(arr, figsize=(4, 4))
show(create_template(41), figsize=(2, 2))
show(hough(arr, 41), figsize=(4, 4))
```

```python
from hough import find_circles
circles = find_circles(arr, 41, 0.99)
print(circles)
```
